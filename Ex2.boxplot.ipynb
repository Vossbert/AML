```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
import pandas as pd
np.random.seed(1)
```

```python
# Not normally distributed
df_nnorm = pd.read_csv("data/Ex1/Coca_Cola_historical_data.csv") 
# First normally distributed
df_norm1 = pd.read_csv("data/Ex1/random_fitness_dataset.csv") 
# Second normally distributed
df_norm2 = pd.read_csv("data/Ex1/SOCR-HeightWeight.csv") 
```

## Dataset Coca Cola 'Close Price'

```python
sns.boxplot(data=df_nnorm['Close'])
plt.show()
```

### Description

Highly skewed distribution, heavily concentrated near the bottom (around 0-20). The box itself is very small compared to the whiskers and the extreme outliers.
based on the concentration on the bottom, you can derive that there has not been significant price movements for a long period.
The outliers above the top whisker represent a recent, significant upward price surge.

## Fitness Dataset 'calories_burned'

```python
sns.boxplot(data=df_norm1['calories_burned'])
plt.show()
```

### Description
A symmetrical and tight distribution. The median is central, and the whiskers are roughly equal in length. There are no visible outliers beyond the whiskers.
Even though the Shapiro-Wilk test just barely got passed with a P-value of 0.05, the boxplot confirms that this is near-normally distributed.

## Height Dataset

```python
sns.boxplot(data=df_norm2['Height(Inches)'])
plt.show()
```

### Description
A symmetrical distribution with a central median. The whiskers are relatively short and equal. There are a number of data points marked as outliers.
The boxplot shows a normal distribution, aligning with the histogram of exercise 1. Median is central and the data is concentrated around the mean.
There are still numerous outliers, but they can be seen as natural variations in a real-world normal distribution, rather than actual errors.
